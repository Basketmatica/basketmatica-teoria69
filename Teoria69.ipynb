# 1. Teoría del 69

```python
# Empezamos por Euroliga
df_puntos_euroleague['Cumple Teoría 69'] = False
cumple_teoria69_euroleague = 0
for game_id, group in df_puntos_euroleague.groupby('game_id'):
    # Variable para monitorear el primer equipo en alcanzar los 69 puntos
    equipo_69 = None
    
    # Iterar por cada jugada del partido
    for _, row in group.iterrows():
        if row['points_a'] >= 69 or row['points_b'] >= 69:
            equipo_69 = 'A' if row['points_a'] >= 69 else 'B'
            break
            
    # Determinar el equipo ganador según los puntos tras el último tiro del partido
    ganador = 'A' if group.iloc[-1]['points_a'] > group.iloc[-1]['points_b'] else 'B'
    
    # Verificar si el ganador coincide con el equipo que anotó primero
    cumple_teoria = equipo_69 == ganador

    # Actualizar la columna en el DataFrame solo para las filas del partido actual
    df_puntos_euroleague.loc[df_puntos_euroleague['game_id'] == game_id, 'Cumple Teoría 69'] = cumple_teoria

    # Incrementar el contador si la teoría se cumple
    if cumple_teoria:
        cumple_teoria69_euroleague += 1


# Cálculo del porcentaje
porcentaje_69_euroleague = cumple_teoria69_euroleague / total_partidos_euroleague * 100
print(f"Porcentaje de éxito de la Teoría del 69 en Euroliga: {porcentaje_69_euroleague:.2f}%")
```

# 4. Análisis de la Teoría de Jugadores con 10+ Puntos

Antes de comenzar con el análisis de la teoría, es necesario realizar algunos pasos para preparar los datos:

1. Asignar los equipos team_a y team_b en el dataframe de puntos (df_puntos) según los cambios en el marcador (points_a y points_b).
2. Vincular los identificadores reales de los equipos (team_id) con team_a y team_b para identificarlos claramente.
3. Combinar el dataframe de puntos con el boxscore (df_boxscore) para relacionar las estadísticas de jugadores con la información del equipo y el resultado del partido.

```python
# Crear un DataFrame para almacenar los equipos asociados a cada partido
equipos_asociados = {}

# Iterar por cada partido
for game_id, group in df_puntos_euroleague.groupby('game_id'):
    team_a, team_b = None, None
    prev_points_a, prev_points_b = 0, 0  # Puntos iniciales
    for _, row in group.iterrows():
        # Si cambia points_a, asociamos ese team_id a team_a
        if row['points_a'] != prev_points_a and team_a is None:
            team_a = row['team_id']
        # Si cambia points_b, asociamos ese team_id a team_b
        if row['points_b'] != prev_points_b and team_b is None:
            team_b = row['team_id']

        # Actualizamos los valores previos
        prev_points_a = row['points_a']
        prev_points_b = row['points_b']

        # Si ya tenemos ambos equipos, podemos detenernos
        if team_a and team_b:
            break

    # Guardamos los equipos para el partido actual
    equipos_asociados[game_id] = {'team_a': team_a, 'team_b': team_b}

# Convertir el diccionario en un DataFrame
equipos_df = pd.DataFrame.from_dict(equipos_asociados, orient='index').reset_index()
equipos_df.columns = ['game_id', 'team_a', 'team_b']

# Unir con el marcador final
df_resultados_euroleague = df_puntos_euroleague.groupby('game_id').tail(1).merge(
    equipos_df, on='game_id'
)

# Agregar los puntos finales de cada equipo
df_resultados_euroleague['points_team_a'] = df_resultados_euroleague['points_a']
df_resultados_euroleague['points_team_b'] = df_resultados_euroleague['points_b']


df_resultados_euroleague.tail()
        
```

```python
# Juntar el dataframe resultante con el boxscore
df_conjunto_euroleague = df_boxscore_euroleague.merge(
    df_resultados_euroleague[['game_id', 'team_a', 'team_b', 'points_team_a', 'points_team_b']],
    on='game_id'
)
df_conjunto_euroleague.head()
```

Tras hacerlo para ambas competiciones y tener el dataframe preparado, vamos con la teoría

```python
# Inicializamos el contador para la teoría
df_conjunto_euroleague['Cumple Teoría Jugadores'] = False
cumple_teoriajugadores_euroleague = 0

# Iteramos por cada partido
for game_id, group in df_conjunto_euroleague.groupby('game_id'):
     # Contadores para jugadores que alcanzan 10 puntos por equipo
    jugadores_a, jugadores_b = 0, 0
    
    # Contamos cuántos jugadores alcanzaron 10 puntos o más en cada equipo
    for _, row in group.iterrows():
        if row['team_id'] == group['team_a'].iloc[0] and row['points'] >= 10:
            jugadores_a += 1
        elif row['team_id'] == group['team_b'].iloc[0] and row['points'] >= 10:
            jugadores_b += 1

    # Determinar el equipo ganador
    ganador = 'A' if group['points_team_a'].iloc[0] > group['points_team_b'].iloc[0] else 'B'

    # Verificar si el equipo con más jugadores con 10+ puntos ganó el partido
    cumple_teoria = (jugadores_a > jugadores_b and ganador == 'A') or (jugadores_b > jugadores_a and ganador == 'B')

    # Actualizar la columna para las filas del partido actual
    df_conjunto_euroleague.loc[df_conjunto_euroleague['game_id'] == game_id, 'Cumple Teoría Jugadores'] = cumple_teoria

    # Incrementar el contador si la teoría se cumple
    if cumple_teoria:
        cumple_teoriajugadores_euroleague += 1

# Calcular el porcentaje de éxito de la teoría
porcentaje_jugadores_euroleague = (cumple_teoriajugadores_euroleague / total_partidos_euroleague) * 100
print(f"Porcentaje de éxito de la Teoría de Jugadores con 10+ Puntos en Euroliga: {porcentaje_jugadores_euroleague:.2f}%")
```
